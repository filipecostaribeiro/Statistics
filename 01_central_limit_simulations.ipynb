# Central Limit Theorem Simulations

## Central Limit Theorem
It states that given any distribution with mean ($\mu$) and standard deviation ($\sigma$), if you collect a large number of samples with size $n$ and take the mean for each sample, the distribution of those means will tend to a normal distribution with mean ($\mu$) and standard deviation ($\frac{\sigma}{\sqrt{n}}$)

## Simulation

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
import random
```

### Exponential distribution

```python
# mean and population size
mu = 2
pop_size = 10000

#plot
pop_exponential = np.random.exponential(mu,pop_size).tolist()
plt.hist(x=pop_exponential,bins = 'auto', rwidth = 0.85)
plt.show()
```

```python
# define the population to work on
population = pop_exponential
# sample size and number of samples
sample_size = [1,2,5,10,50,400]
n_samples = 5000

# list to store the mean for each sample

# extraction of mean and standard deviation from the population
population_mean = np.mean(population)
std_population = np.std(population)

fig, axes = plt.subplots(3,2, figsize=(15,10))
axes = axes.flatten()

for a in range(0,len(sample_size)):
    n_sample_size = sample_size[a]
    list_mean = []
    
    # beginning of the simulation
    for i in range(0,n_samples):
        X = np.random.choice(population, n_sample_size,replace = True)
        list_mean.append(np.mean(X))
    
    # histogram plot for the means
    axes[a].hist(x=list_mean, bins = 'auto', rwidth = 0.85,density=True, label=f'sample size:{n_sample_size}')
    # Plot of normal distribution with the mean and standard deviation given by the Central Limit Theorem.
    xmin, xmax = min(list_mean), max(list_mean)
    x = np.linspace(xmin, xmax, len(list_mean))
    y = norm.pdf(x, population_mean, (std_population / np.sqrt(n_sample_size)))
    axes[a].plot(x, y)

    axes[a].legend()
plt.show()
```

### Uniform distribution

```python
low = 0  # Limite inferior
high = 10  # Limite superior
pop_size = 10000
uniform_dist = np.random.uniform(low, high, pop_size).tolist()
plt.hist(x=uniform_dist,bins = 'auto', rwidth = 0.85)
plt.show()
```

```python
# define the population to work on
population = uniform_dist
# sample size and number of samples
sample_size = [1,2,5,10,50,400]
n_samples = 5000

# list to store the mean for each sample

# extraction of mean and standard deviation from the population
population_mean = np.mean(population)
std_population = np.std(population)

fig, axes = plt.subplots(3,2, figsize=(15,10))
axes = axes.flatten()

for a in range(0,len(sample_size)):
    n_sample_size = sample_size[a]
    list_mean = []
    
    # beginning of the simulation
    for i in range(0,n_samples):
        X = np.random.choice(population, n_sample_size,replace = True)
        list_mean.append(np.mean(X))
    
    # histogram plot for the means
    axes[a].hist(x=list_mean, bins = 'auto', rwidth = 0.85,density=True, label=f'sample size:{n_sample_size}')
    # Plot of normal distribution with the mean and standard deviation given by the Central Limit Theorem.
    xmin, xmax = min(list_mean), max(list_mean)
    x = np.linspace(xmin, xmax, len(list_mean))
    y = norm.pdf(x, population_mean, (std_population / np.sqrt(n_sample_size)))
    axes[a].plot(x, y)

    axes[a].legend()
plt.show()
```
